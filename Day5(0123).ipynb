3. 학생의 이름/국어점수/영어점수/수학점수를 입력받아 Pandas의 Dataframe으로 구성하는 함수

```python
import pandas as pd
```

```python
pd.DataFrame(data={'이름':['홍길동','성춘향'],
                  '국어':[100,90], '영어':[80,70],'수학':[50,60]})
```

```python
def func3():
    # 빈 리스트를 초기화
    name_list = []
    score_list1 =[]
    score_list2 = []
    score_list3 = []
    #학생 수 입력 받기
    stu_num = int(input('학생 수를 입력하시오 : '))
    for i in range(1, stu_num+1):
        name = input('번째 학생의 이름을 입력하시오.')
        s1 = input('국어 점수를 입력하시오: ')
        s2 = input('영어 점수를 입력하시오: ')
        s3 = input('수학 점수를 입력하시오: ')
        name_list.append(name)
        score_list1.append(s1)
        score_list2.append(s2)
        score_list3.append(s3)
    df_stu = pd.DataFrame(data={'이름':name_list, '국어':score_list1, '영어':score_list2,'수학': score_list3})
    return df_stu
```

# 기술 통계량 확인
- 데이터의 통계적 특성을 확인하기 위해
- 연속형 :
    1. 대표값 (평균 mean / 중위수 median) : 해당 연속형 자료를 대표하는 대표값
        - 평균의 경우 이상치(outlier, 데이터 트랜드에 벗어나는 값)에 의해 매우 높게 계산될 수 있음
        A : 1, 4, 2, 3, 5 / 평균 : 3 , 중앙값 : 3 
        B : 1, 4, 2, 3, 1000 / 평균 : 약 200 , 중앙값 : 3
        
    2. 산포 (분산 / 표준편차) : 해당 연속형 자료의 대표값으로 부터 데이터들이 얼마나 떨어져 있는지 나타내는 통계량
        - 대푯값에 대해 데이터나 얼마나 정확한가를 나타내는 척도
        - 데이터의 신뢰성을 나타내기도 함
        
    3. 분포의 모양 : 해당 연속형 자료의 분포모양
        - 분포의 모양에 따라 가설검정 기법 / 전통통계에서 사용하는 회귀분석 기법이 달라질 수 있기 때문!
        - 가설 검정을 통해 확인 (정규성 검정)

- 범주형 : 항목 / 빈도수

# 데이터 전처리

- 층별화 : 범주형 데이터의 항목의 특성에 따라 다른 데이터의 통계적 특성이 달라짐
- 데이터 필터 (Filter) 특정 조건에 따라 데이터를 나누어 확인/처리

```python
# 월랜탈비용이 10만원 이상인 계약 명단을 확인
cond1 = (df1['Amount_Month'] >= 100000)
df1.loc[cond1]   #loc (location) : Series에서 True 값으로 잡힌 Index에 해당하는 데이터를 추출
```

```python
# df1[cond1]['Channel']
```

```python
# 월랜탈비용은 10만원 이상이고, 연체건수는 2건 미만인 계약 명단을 추출
cond1 = (df1['Amount_Month']>=100000)
cond2 = (df1['Overdue_count']<2)
df1.loc[cond1 & cond2] #위 두 조건을 모두 만족하는 데이터를 추출 (And)
```

```python
# 두 조건 중 하나의 조건만 만족해도 데이터를 추출
df1.loc[cond1|cond2]
```

```python
# 판매유형(Sales_Type)이 '렌탈'에 해당하는 데이터만 추출 (특정 항목값만 추출)
cond1 = (df1['Sales_Type']=='렌탈')
df1.loc[cond1]
```

```python
# 유입경로가 '대형마트H', '대형마트E', '대형마트A', '대형마트N', '대형마트C'

# isin() : 특정 리스트에 값이 있다면 True 없으면 False
store_list = ['대형마트H', '대형마트E', '대형마트A', '대형마트N', '대형마트C']
cond1 = df1['Channel'].isin(store_list)
df1.loc[cond1]
```

```python
# 사용자가 직접 만든 함수에다가 데이터를 처리 할 때
def func1(row):
    if '대형마트' in row:
        return True
    else:
        return False
    
# apply( ) : 데이터 (시리즈)내 데이터 한 행씩, 직접 만든 함수에 적용
cond1 = df1['Channel'].apply(func1)
df1.loc[cond1]
```

## **날짜 데이터 처리**

```python
# 날짜 데이터를 처리 (Python 날짜 형식의 데이터 타입으로 변환)
df1['Datetime_dt'] = pd.to_datetime(df1['Datetime'])
df1.head(2)
```

```python
# 날짜 또는 시간 추출
df1['Year'] = df1['Datetime_dt'].dt.year # 연도를 추출
df1['Month'] = df1['Datetime_dt'].dt.month # 월 추출
df1['Day'] = df1['Datetime_dt'].dt.day # 일 추출
df1['Day_of_week'] = df1['Datetime_dt'].dt.day_name() # 요일 추출
df1['Week'] = df1['Datetime_dt'].dt.isocalendar().week # 주차 (1주차 ~ 52주차)
```

## **파생 변수 생성**
: 기존의 항목을 이용해 계산하여 만든 새로운 항목

```python
# 단순 사칙연산을 이용한 파생 변수 생성
df1['Total_Amount'] = df1['Term'] * df1['Amount_Month']
df1['Total_Amount'].describe()
```

```python
# 연속형 자료를 범주형으로 그룹화
cond1 = (df1['Age']>=70)
cond2 = (df1['Age']>=60)
cond3 = (df1['Age']>=50)
cond4 = (df1['Age']>=40)
cond5 = (df1['Age']<40)

df1.loc[cond5 , 'Age_group'] = '30대이하'
df1.loc[cond4 , 'Age_group'] = '40대이상'
df1.loc[cond3 , 'Age_group'] = '50대이상'
df1.loc[cond2 , 'Age_group'] = '60대이상'
df1.loc[cond1 , 'Age_group'] = '70대이상'
```

```python
df1['Age_group'].value_counts()
```

```python
# Apply 기법을 이용하여
def func2(row):
    if pd.isnull(row): # 입력받은 값이 결측치이면 (결측값이면 True)
        return np.nan # 결측값 그대로 내버려 둠
    
    if row >= 100:
        return '초고령'
    
    else:
        return str(row)[0] + '0대'
    
df1['연령대'] = df1['Age'].apply(func2)
df1['연령대'].value_counts()
```

**결측값 처리**

- 결측값 : 데이터를 수집하거나 처리하는 과정에서 오류나 누락으로 인해 발생한 공백
- 결측값 처리 기법:
    - 결측값 제거 : 결측값이 존재하는 행을 제거
    - 결측값 대치 : 결측값을 다른 값으로 채워넣는 기법
        - 단순대치 : 특저 지정된 값(대표값 또는 상수) 대치
            -> 항목들 간의 여관성이나 상관성이 명확할 때
            -> 사회과학 데이터 / 병원 / ...
        - 알고리즘 대치 : KNN과 같은 머신러닝 알고리즘을 이용해 대치
        - 보간법 : 특정 간격 사이에 결측치가 있을 때, 간격의 크기대로 결측값을 대치
